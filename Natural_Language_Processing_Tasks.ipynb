# Μέρος Α: Fine-tune a pretrained model

Τα γλωσσικά μοντέλα αποτελούνται από δύο στάδια εκπαίδευσης:
1. Pre-training σε μεγάλα unlabelled datasets. Το pre-training είναι υπολογιστικά πολύ ακριβό και γι αυτό στην πράξη δε το χρησιμοποιούμε όταν θέλουμε να τρέξουμε ένα μοντέλο σε ένα καινούργιο dataset. Μπορούμε να σκεφτούμε το pre-training ως τη διαδικασία εκμάθησης γλωσσικών κανόνων κι εννοιών, οι οποίες στη συνέχεια μπορούν να χρησιμοποιηθούν για διάφορους σκοπούς.

2. Fine-tuning σε μικρότερα labelled datasets. Το fine-tuning πρακτικά εκμεταλλεύεται τις ιδιότητες του transfer learning προκειμένου να μεταφέρουμε τη 'γνώση' που έχει αποθηκευθεί στο γλωσσικό μοντέλο κατά τη διάρκεια του pre-training σε συγκεκριμένα task. Κάθε task εξυπηρετείται μέσω στοχευμένων datasets. Για παράδειγμα, κάποια datasets αναφέρονται στην ταξινόμηση κειμένων σε κατηγιορίες (text classification), άλλα datasets περιέχουν ερωτήσεις οι οποίες πρέπει να απαντηθούν (question answering) κι άλλα πολλά.

Κάποια κλασικά tasks της επεξεργασίας φυσικής γλώσσας είναι τα ακόλουθα:
- Text classification
- Question answering
- Natural language inference
- Fill mask
- Semantic similarity

Περισσότερες πληροφορίες μπορείτε να βρείτε στον ακόλουθο σύνδεσμο στο domain Natural Language Processing: https://huggingface.co/models

**Σημείωση Optimizers:**

Εκτός από την αρχικοποίηση, η επιλογή ενός κατάλληλου αλγορίθμου βελτιστοποίησης είναι σημαντική για τα βαθιά νευρωνικά δίκτυα.

Ο βελτιστοποιητής είναι υπεύθυνος για την ενημέρωση των παραμέτρων του δικτύου  δεδομένου των κλίσεων.

Ως εκ τούτου, υλοποιούμε αποτελεσματικά μια συνάρτηση $w^{t} = f(w^{t-1}, g^{t}, ...)$ με τις παραμέτρους $w$ και $g^{t} = \nabla_{w^{(t-1)}} \mathcal{L}^{(t)}$ οι κλίσεις στο χρονικό βήμα $t$.

Μια κοινή, πρόσθετη παράμετρος σε αυτή τη συνάρτηση είναι ο ρυθμός εκμάθησης (learning rate), που εδώ υποδηλώνεται με $\eta$. Συνήθως, ο ρυθμός εκμάθησης μπορεί να θεωρηθεί ως το "μέγεθος βήματος" της ενημέρωσης (update). Ένας υψηλότερος ρυθμός μάθησης σημαίνει ότι αλλάζουμε τα βάρη περισσότερο προς την κατεύθυνση των κλίσεων, ένας μικρότερος σημαίνει ότι κάνουμε πιο σύντομα βήματα.
Καθώς οι περισσότεροι βελτιστοποιητές διαφέρουν μόνο στην εφαρμογή του $f$, μπορούμε να ορίσουμε ένα template για έναν βελτιστοποιητή στο PyTorch παρακάτω.

Λαμβάνουμε ως είσοδο τις παραμέτρους ενός μοντέλου και έναν ρυθμό εκμάθησης. Η συνάρτηση `zero_grad` μηδενίζει τις κλίσεις όλων των παραμέτρων, κάτι που πρέπει να κάνουμε πριν καλέσουμε το `loss.backward()`. Τέλος, η συνάρτηση  `step()` λέει στον βελτιστοποιητή να ενημερώσει όλα τα βάρη με βάση τις κλίσεις τους.



---


Βασικό εργαλείο βελτιστοποίησης αποτελεί το τυπικό **Stochastic Gradient Descent (SGD)**. Το SGD ενημερώνει τις παραμέτρους χρησιμοποιώντας την ακόλουθη εξίσωση:

$$
\begin{split}
    w^{(t)} & = w^{(t-1)} - \eta \cdot g^{(t)}
\end{split}
$$

Χρησιμοποιώντας την έννοια momentum αντικαθιστούμε την κλίση στην ενημέρωση με έναν εκθετικό μέσο όρο όλων των προηγούμενων κλίσεων συμπεριλαμβανομένης της τρέχουσας:

$$
\begin{split}
    m^{(t)} & = \beta_1 m^{(t-1)} + (1 - \beta_1)\cdot g^{(t)}\\
    w^{(t)} & = w^{(t-1)} - \eta \cdot m^{(t)}\\
\end{split}
$$


**Σημείωση:**

Ο **Adam** συνδυάζει την ιδέα της momentum με έναν προσαρμοστικό ρυθμό μάθησης (adaptive learning rate), ο οποίος βασίζεται σε έναν εκθετικό μέσο όρο των τετραγωνικών κλίσεων, δηλαδή στη νόρμα των κλίσεων.

Επιπλέον, προσθέτουμε μια διόρθωση στο bias για τo momentum και το adaptive learning rate για τις πρώτες επαναλήψεις:

$$
\begin{split}
    m^{(t)} & = \beta_1 m^{(t-1)} + (1 - \beta_1)\cdot g^{(t)}\\
    v^{(t)} & = \beta_2 v^{(t-1)} + (1 - \beta_2)\cdot \left(g^{(t)}\right)^2\\
    \hat{m}^{(t)} & = \frac{m^{(t)}}{1-\beta^{t}_1}, \hat{v}^{(t)} = \frac{v^{(t)}}{1-\beta^{t}_2}\\
    w^{(t)} & = w^{(t-1)} - \frac{\eta}{\sqrt{\hat{v}^{(t)}} + \epsilon}\circ \hat{m}^{(t)}\\
\end{split}
$$

Το Epsilon είναι μια μικρή σταθερά που χρησιμοποιείται για τη βελτίωση της αριθμητικής σταθερότητας για πολύ μικρές κλίσεις. Θυμηθείτε ότι ο προσαρμοστικός ρυθμός εκμάθησης δεν αντικαθιστά την υπερπαράμετρο ρυθμού εκμάθησης $\eta$, αλλά μάλλον λειτουργεί ως επιπλέον παράγοντας και διασφαλίζει ότι οι κλίσεις των διαφόρων παραμέτρων έχουν παρόμοια νόρμα.


Τέλος η βελτιστοποίηση **AdamW** είναι μια μέθοδος στοχαστικής κλίσης καθόδου που βασίζεται στην προσαρμοστική εκτίμηση ροπών πρώτης και δεύτερης τάξης με μια πρόσθετη μέθοδο στα βάρη αποσύνθεσης (weight decay).

**Σημείωση Matthew Correlation metric:**

Κατά τη διάρκεια της έρευνας σχετικά με την παρούσα εργασία, εντοπίσαμε τον εξής συντελεστή συσχέτισης ως προτεινόμενη μετρική αξιολόγησης. Βέβαια ο Main κώδικας της εργασία χρησιμοποιεί ως μετρική το accuracy οπότε παραμείναμε σε αυτή την γραμμή.

Η ακρίβεια είναι ευαίσθητη στην ανισορροπία κατηγορίας. Η ακρίβεια - precision, η ανάκληση - recall και η F1-score είναι ασύμμετρες. Τι πρέπει λοιπόν να κάνει κανείς; Εάν ενδιαφέρουν και οι δύο κατηγορίες, μπορεί κανείς να αντιμετωπίσει ένα πρόβλημα δυαδικής ταξινόμησης ως πρόβλημα πολλαπλών κλάσεων με 2 κλάσεις και στη συνέχεια να υπολογίσει τις αντίστοιχες μετρήσεις πολλαπλών κλάσεων: micro- or macro-averaged precision, recall, and F1-score.

Για τη δυαδική ταξινόμηση, υπάρχει μια άλλη (και αναμφισβήτητα πιο κομψή).

Λύση: Αντιμετώπιση της αληθινής τάξης και της προβλεπόμενης τάξης ως δύο (δυαδικές) μεταβλητές και μετέπειτα υπολογισμός του συντελεστή συσχέτισής τους (με παρόμοιο τρόπο με τον υπολογισμό του συντελεστή συσχέτισης μεταξύ οποιωνδήποτε δύο μεταβλητών). Όσο υψηλότερη είναι η συσχέτιση μεταξύ αληθινών και προβλεπόμενων τιμών, τόσο καλύτερη είναι η πρόβλεψη. Ο συντελεστής συσχέτισης Matthews (MCC) όταν εφαρμόζεται σε ταξινομητές ορίζεται ως:

<center>
$MCC = \frac{TP \hspace{2mm} x \hspace{2mm} TN \hspace{2mm} - \hspace{2mm} FP  \hspace{2mm} x \hspace{2mm}FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}$
</center>


Μερικές ωραίες ιδιότητες του MCC μπορούν εύκολα να εξαχθούν από αυτόν τον τύπο: όταν ο ταξινομητής είναι τέλειος (FP = FN = 0), η τιμή του MCC είναι 1, υποδεικνύοντας τέλεια θετική συσχέτιση.


Αντίθετα, όταν ο ταξινομητής κάνει πάντα εσφαλμένη ταξινόμηση (TP = TN = 0), παίρνουμε μια τιμή -1, που αντιπροσωπεύει τέλεια αρνητική συσχέτιση (σε αυτήν την περίπτωση, μπορείτε απλά να αντιστρέψετε το αποτέλεσμα του ταξινομητή για να πάρετε τον ιδανικό ταξινομητή).

Στην πραγματικότητα, η τιμή MCC είναι πάντα μεταξύ -1 και 1, με το 0 να σημαίνει ότι ο ταξινομητής δεν είναι καλύτερος από μια τυχαία ανατροπή ενός εύλογου νομίσματος. Το MCC είναι επίσης απόλυτα συμμετρικό, επομένως καμία τάξη δεν είναι πιο σημαντική από την άλλη. Αν αλλάξουμε το θετικό και το αρνητικό, θα εξακολουθούμε να έχετε την ίδια τιμή.

Source:  https://towardsdatascience.com/the-best-classification-metric-youve-never-heard-of-the-matthews-correlation-coefficient-3bf50a2f3e9a

Μερικές παραδοχές που έπρεπε να γίνουν ώστε να συμπέσει το task με το δοσμένο dataset είναι:


*   To πεδίο goal είναι η πρόταση που δίνεται στο μοντέλο
*   Στο μοντέλο παρέχονται μόνο 2 επιλογές, τα πεδία sol1, sol2

Τα παρακάνω μοντέλα επιλέχτηκαν από το hugging face (https://huggingface.co/models?other=multiple-choice) και έχουν ως βάση fine-tuned μοντέλα στο task του mulitple-choice με την χρήση διαφορετικών datasets.

````python
# insert your code here (models)

roberta_tokenizer = AutoTokenizer.from_pretrained('Riiid/kda-roberta-base-race')
roberta_model = AutoModelForMultipleChoice.from_pretrained('Riiid/kda-roberta-base-race', return_dict=True)

bert_swag_tokenizer = AutoTokenizer.from_pretrained('aaraki/bert-base-uncased-finetuned-swag')
bert_swag_model = AutoModelForMultipleChoice.from_pretrained('aaraki/bert-base-uncased-finetuned-swag', return_dict=True)

bert_wsc_tokenizer = AutoTokenizer.from_pretrained('nightingal3/bert-finetuned-wsc')
bert_wsc_model = AutoModelForMultipleChoice.from_pretrained('nightingal3/bert-finetuned-wsc', return_dict=True)

distil_roberta_tokenizer = AutoTokenizer.from_pretrained('gsgoncalves/distilroberta-base-race')
distil_roberta_model = AutoModelForMultipleChoice.from_pretrained('gsgoncalves/distilroberta-base-race', return_dict=True)

albert_tokenizer = AutoTokenizer.from_pretrained('jeremykke/albert-base-v2-finetuned-swag')
albert_model = AutoModelForMultipleChoice.from_pretrained('jeremykke/albert-base-v2-finetuned-swag', return_dict=True)
````

````python
# insert your code here (function for ending prediction)

def prediction(dataset, model, tokenizer):
    true_labels = []
    predicted_labels = []
    for idx in dataset.index:

      # For every element in dataset get question and the two choices
      prompt = dataset["goal"][idx]
      choices = [dataset["sol1"][idx],dataset["sol2"][idx]]
      true_labels.append(dataset["label"][idx])

      # Calculate model's predictions
      model.eval()
      with torch.no_grad():
        encoding = tokenizer([[prompt, choices[0]], [prompt, choices[1]]], return_tensors='pt', padding=True)
        outputs = model(**{k: v.unsqueeze(0) for k,v in encoding.items()}) # batch size is 1
        prob = torch.softmax(outputs.logits, dim=-1)[0].tolist()
        predicted_labels.append(np.argmax(prob))

    # Return accuracy
    return accuracy_score(true_labels,predicted_labels)
````

Η σύγκριση των διανυσματικών αναπαραστάσεων μπορεί να γίνει κλασικά μέσω μεθόδων όπως το consine similarity, με μεγαλύτερες τιμές μεταξύ διανυσμάτων να σηματοδοτούν πιο όμοια διανύσματα, άρα και πιο όμοιες προτάσεις. Δίνουμε για το λόγο αυτό μια συνάρτηση υπολογισμού του cosine similarity.

````python
from sklearn.metrics.pairwise import cosine_similarity
def get_cosine_similarity(feature_vec_1, feature_vec_2):
    return cosine_similarity(feature_vec_1.reshape(1, -1), feature_vec_2.reshape(1, -1))[0][0]
````

````python
# insert your code here (load models, create pipelines)

from transformers import pipeline, AutoTokenizer

# Load model's pipeline and the corresponding tokenizers
distilbert_tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
albert_tokenizer = AutoTokenizer.from_pretrained("albert-base-v2")
roberta_tokenizer = AutoTokenizer.from_pretrained("roberta-base")

fill_masker_distilbert = pipeline(model="distilbert-base-uncased")
fill_masker_albert = pipeline(model="albert-base-v2")
fill_masker_roberta = pipeline(model="roberta-base")
````

````python
# insert your code here (trnasform dataset)

# We need to keep only word options that exist in the vocabulary of all models
sentence_list = []
answer_list = []
option_1_list = []
option_2_list = []
vocab1 = distilbert_tokenizer.get_vocab()
vocab2 = albert_tokenizer.get_vocab()
vocab3 = roberta_tokenizer.get_vocab()
i = 0
for idx in winogrande_df.index:

    # If words from both options exist in the in the vocabulary of all models
    if winogrande_df.iloc[idx]["option1"] in vocab1.keys() and winogrande_df.iloc[idx]["option2"] in vocab1.keys() and winogrande_df.iloc[idx]["option1"] in vocab2.keys() and winogrande_df.iloc[idx]["option2"] in vocab2.keys() and winogrande_df.iloc[idx]["option1"] in vocab3.keys() and winogrande_df.iloc[idx]["option2"] in vocab3.keys():
        sentence_list.append(winogrande_df.iloc[idx]["sentence"])
        answer_list.append(winogrande_df.iloc[idx]["answer"])
        option_1_list.append(winogrande_df.iloc[idx]["option1"])
        option_2_list.append(winogrande_df.iloc[idx]["option2"])
        i = i+1
    if i >= 100: # 100 samples
        break

# Create the new dataset
winogrande_df = pd.DataFrame({'sentence': sentence_list,
        'answer': answer_list,
        'option1': option_1_list,
        'option2': option_2_list})
````

````python
# insert your code here (function for predicting best filling)

def prediction(dataset, fill_masker, mask_token):
    true_labels = []
    predicted_labels = []

    # For every item in dataset
    for idx in dataset.index:

        # Get model's predictions
        targets = [dataset.iloc[idx]["option1"], dataset.iloc[idx]["option2"]]
        sentence = dataset.iloc[idx]["sentence"]
        pred_word_option = fill_masker(sentence.replace("_",mask_token), targets=targets)[0]['token_str'] # Take the more possible word of the two
        predicted_labels.append(targets.index(pred_word_option)+1)
        true_labels.append(int(dataset.iloc[idx]["answer"]))

    # Return accuracy
    return accuracy_score(true_labels,predicted_labels)
````
